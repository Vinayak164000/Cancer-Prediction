```python
df.drop('STDs: Time since first diagnosis', axis = 1, inplace = True)
df.drop('STDs: Time since last diagnosis', axis = 1, inplace = True)
df.drop('Hormonal Contraceptives (years)', axis = 1, inplace = True)
```

```python
def drop_functions(data):
    cols = ['IUD','IUD (years)','STDs','STDs (number)','STDs:condylomatosis','STDs:cervical condylomatosis','STDs:vaginal condylomatosis','STDs:vulvo-perineal condylomatosis','STDs:syphilis','STDs:pelvic inflammatory disease','STDs:genital herpes','STDs:molluscum contagiosum','STDs:AIDS','STDs:HIV','STDs:Hepatitis B','STDs:HPV']
    for i in cols:
        data.drop(i, axis = 1, inplace = True)

drop_functions(df)
```

```python
X = df.drop('Biopsy', axis = 1)
Y = df['Biopsy']
```

```python
X.columns
```

```python
X.info()
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
```

```python
Y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train,Y_train)
```

```python
print(classifier.coef_.shape)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score,accuracy_score, precision_score, recall_score
cm = confusion_matrix(y_pred, Y_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_test, y_pred))
```

```python
accuracy_score(Y_test, y_pred), f1_score(Y_test, y_pred), recall_score(Y_test, y_pred), precision_score(Y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, Y_train)
pred = clf.predict(X_test)
```

```python
cm = confusion_matrix(pred, Y_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report())
```

```python
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier()
clf2.fit(X_train, Y_train)
pred = clf2.predict(X_test)
```

```python
cm = confusion_matrix(pred, Y_test)
sns.heatmap(cm, annot = True)
```
